```python
%matplotlib inline
import pandas
import numpy
import seaborn
import scipy
import matplotlib.pyplot as plt
seaborn.set_context(font_scale=1.5)
```

```python
data=pandas.read_excel('/Users/arjun/Google Drive/HDCTF/Data/170825_ABS_PIN_Phenotyping/170825_number_branches.xlsx')

fig = seaborn.FacetGrid(data=data,row='Background',aspect=1.5,sharex=False,sharey=False,size=5)
fig = fig.map(seaborn.boxplot,'Nem#','Number_branches_including_stems',palette='Greys')
fig = fig.map(seaborn.swarmplot,'Nem#','Number_branches_including_stems','Nem#',palette='Spectral').add_legend()
plt.legend(bbox_to_anchor=(1, 1), loc=2)

fig = seaborn.FacetGrid(data=data,row='Background',aspect=1.5,sharex=False,sharey=False,size=5)
fig = fig.map(seaborn.boxplot,'Nem#','Number of branches on main stem',palette='Greys')
fig = fig.map(seaborn.swarmplot,'Nem#','Number of branches on main stem','Nem#',palette='Spectral')
plt.legend(bbox_to_anchor=(1, 1), loc=2)

fig = seaborn.FacetGrid(data=data,row='Background',aspect=1.5,sharex=False,sharey=False,size=5)
fig = fig.map(seaborn.boxplot,'Nem#','Number_stems',palette='Greys')
fig = fig.map(seaborn.swarmplot,'Nem#','Number_stems','Nem#',palette='Spectral')
plt.legend(bbox_to_anchor=(1, 1), loc=2)


#looking at anova
anova = scipy.stats.f_oneway(data[data['Nem#']==4148]['Number_branches_including_stems'],data[data['Nem#']==3811]['Number_branches_including_stems'])
print('Anova 4148 vs 3811 (Number of branches including stems): '+str(anova))
anova = scipy.stats.f_oneway(data[data['Nem#']==4150]['Number_branches_including_stems'],data[data['Nem#']==3811]['Number_branches_including_stems'])
print('Anova 4150 vs 3811 (Number of branches including stems): '+str(anova))
anova = scipy.stats.f_oneway(data[data['Nem#']==4152]['Number_branches_including_stems'],data[data['Nem#']==3811]['Number_branches_including_stems'])
print('Anova 4152 vs 3811 (Number of branches including stems): '+str(anova))

anova = scipy.stats.f_oneway(data[data['Nem#']==4148]['Number of branches on main stem'],data[data['Nem#']==3811]['Number of branches on main stem'])
print('Anova 4148 vs 3811 (Number of branches on main stem): '+str(anova))
anova = scipy.stats.f_oneway(data[data['Nem#']==4150]['Number of branches on main stem'],data[data['Nem#']==3811]['Number of branches on main stem'])
print('Anova 4150 vs 3811 (Number of branches on main stem): '+str(anova))
anova = scipy.stats.f_oneway(data[data['Nem#']==4152]['Number of branches on main stem'],data[data['Nem#']==3811]['Number of branches on main stem'])
print('Anova 4152 vs 3811 (Number of branches on main stem): '+str(anova))
```
