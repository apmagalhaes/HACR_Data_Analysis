```python
%matplotlib inline
import pandas
import scipy
import numpy
import seaborn
import matplotlib.pyplot as plt
seaborn.set_context(font_scale=1.5)
```

```python
data=pandas.read_excel('/Users/arjun/Google Drive/HDCTF/Data/170630_ABS44_PIN1_phenotyping/170630_ABS44_plus_Pin1_number_of_branches.xlsx')
data['Number_of_branches_per_stem'] = [float(x-y)/y for x,y in zip(data['Number_branches_including_stems'],data['Number_stems'])]
```

```python
#plotting data for ABS44 - 3811
data_3811 = data[data['Background']==3811]

fig = seaborn.FacetGrid(data=data_3811,aspect=1.5,sharex=False,sharey=False,size=5)
fig = fig.map(seaborn.boxplot,'Nem#','Number_branches_including_stems',palette='Greys')
fig = fig.map(seaborn.swarmplot,'Nem#','Number_branches_including_stems','Nem#',palette='YlOrRd')

anova_4148 = scipy.stats.f_oneway(data_3811[data_3811['Nem#']==4148]['Number_branches_including_stems'],data_3811[data_3811['Nem#']==3811]['Number_branches_including_stems'])
print(anova_4148)
anova_4150 = scipy.stats.f_oneway(data_3811[data_3811['Nem#']==4150]['Number_branches_including_stems'],data_3811[data_3811['Nem#']==3811]['Number_branches_including_stems'])
print(anova_4150)
anova_4152 = scipy.stats.f_oneway(data_3811[data_3811['Nem#']==4152]['Number_branches_including_stems'],data_3811[data_3811['Nem#']==3811]['Number_branches_including_stems'])
print(anova_4152)
```
