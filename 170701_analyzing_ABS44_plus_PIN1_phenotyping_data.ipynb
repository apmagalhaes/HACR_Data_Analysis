```python
%matplotlib inline
import pandas
import scipy
import numpy
import seaborn
import matplotlib.pyplot as plt
seaborn.set_context(font_scale=1.5)
```

```python
data=pandas.read_excel('/Users/arjun/Google Drive/HDCTF/Data/170630_ABS44_PIN1_phenotyping/170630_ABS44_plus_Pin1_number_of_branches.xlsx')

data['Number_of_branches_per_stem'] = [float(x-y)/y for x,y in zip(data['Number_branches_including_stems'],data['Number_stems'])]
# data=data[data['Genotype']!='Col0']

fig = seaborn.FacetGrid(data=data,col='Background',aspect=1.5,sharex=False,sharey=False,size=5)
fig = fig.map(seaborn.boxplot,'Nem#','Number_stems',palette='Greys')
fig = fig.map(seaborn.swarmplot,'Nem#','Number_stems','Days_since_planting',palette='YlOrRd')
plt.legend(bbox_to_anchor=(1, 1), loc=2)

fig = seaborn.FacetGrid(data=data,col='Background',aspect=1.5,sharex=False,sharey=False,size=5)
fig = fig.map(seaborn.boxplot,'Nem#','Number_branches_including_stems',palette='Greys')
fig = fig.map(seaborn.swarmplot,'Nem#','Number_branches_including_stems','Days_since_planting',palette='YlOrRd')
plt.legend(bbox_to_anchor=(1, 1), loc=2)


#removing the plant lines that had a full or partial sterile phenotype. 
data= data[(data['Sterile']!='sterile')&(data['Sterile']!='early sterile')]

fig = seaborn.FacetGrid(data=data,col='Background',aspect=1.5,sharex=False,sharey=False,size=5)
fig = fig.map(seaborn.boxplot,'Nem#','Number_stems',palette='Greys')
fig = fig.map(seaborn.swarmplot,'Nem#','Number_stems','Days_since_planting',palette='YlOrRd')
plt.legend(bbox_to_anchor=(1, 1), loc=2)

fig = seaborn.FacetGrid(data=data,col='Background',aspect=1.5,sharex=False,sharey=False,size=5)
fig = fig.map(seaborn.boxplot,'Nem#','Number_branches_including_stems',palette='Greys')
fig = fig.map(seaborn.swarmplot,'Nem#','Number_branches_including_stems','Days_since_planting',palette='YlOrRd')
plt.legend(bbox_to_anchor=(1, 1), loc=2)

# fig = seaborn.FacetGrid(data=data,col='Background',aspect=1.5,sharex=False,sharey=False,size=5)
# plt.suptitle('Number stems + branches',size=20)
# fig = fig.map(seaborn.boxplot,'Nem#','Number_of_branches_per_stem',palette='Greys')
# fig = fig.map(seaborn.swarmplot,'Nem#','Number_of_branches_per_stem','Days_since_planting',palette='YlOrRd')
```
