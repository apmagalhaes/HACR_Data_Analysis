```python
%matplotlib inline
import pandas
import numpy
import seaborn
import scipy
import matplotlib.pyplot as plt
seaborn.set_context(font_scale=1.5)
```

```python
data=pandas.read_excel('/Users/arjun/Google Drive/HDCTF/Data/180402 - AuxinHACR PIN1 phenotyping/180402 - AuxinHACR phenotyping.xlsx')

fig = seaborn.FacetGrid(data=data,sharex=False,size=5)
fig = fig.map(seaborn.boxplot,'Nem#','Branch#',palette='Greys')
fig = fig.map(seaborn.swarmplot,'Nem#','Branch#','Genotype',palette='YlOrRd')
plt.legend(bbox_to_anchor=(1, 1), loc=2)

# #looking at anova
anova = scipy.stats.f_oneway(data[data['Nem#']=='4158']['Branch#'],data[data['Nem#']=='3824']['Branch#'])
print('Anova 4158 vs 3824 (Number of branches): '+str(anova))
anova = scipy.stats.f_oneway(data[data['Nem#']=='4159']['Branch#'],data[data['Nem#']=='3824']['Branch#'])
print('Anova 4159 vs 3824 (Number of branches): '+str(anova))
anova = scipy.stats.f_oneway(data[data['Nem#']=='4160']['Branch#'],data[data['Nem#']=='3824']['Branch#'])
print('Anova 4160 vs 3824 (Number of branches): '+str(anova))


```
