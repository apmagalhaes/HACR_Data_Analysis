```python
%matplotlib inline
import pandas
import numpy
import seaborn
import scipy
import matplotlib.pyplot as plt
seaborn.set_context(font_scale=1.5)
```

```python
inter_coin = pandas.read_excel('/Users/arjun/Google Drive/HDCTF/Data/170713_ABS_PIN1_phenotyping/Number_coinitiations_170713.xlsx')
inter_coin['genotype']= ['abs44_3824']*5 + ['abs44+PIN1_4160']*4 + ['abs44+PIN1_4162']*4 + ['abs44+PIN1_4158']*4

plt.figure()
seaborn.boxplot(data=inter_coin,x='genotype',y='Number_coinitations',palette='Greys')
seaborn.swarmplot(data=inter_coin,x='genotype',y='Number_coinitations',hue='Stem_ID',palette='YlOrRd')
plt.legend(bbox_to_anchor=(1, 1), loc=2)

anova_4158 = scipy.stats.f_oneway(inter_coin[inter_coin['genotype']=='abs44+PIN1_4158']['Number_coinitations'],inter_coin[inter_coin['genotype']=='abs44_3824']['Number_coinitations'])
print(anova_4158)
anova_4160 = scipy.stats.f_oneway(inter_coin[inter_coin['genotype']=='abs44+PIN1_4160']['Number_coinitations'],inter_coin[inter_coin['genotype']=='abs44_3824']['Number_coinitations'])
print(anova_4160)
anova_4162 = scipy.stats.f_oneway(inter_coin[inter_coin['genotype']=='abs44+PIN1_4162']['Number_coinitations'],inter_coin[inter_coin['genotype']=='abs44_3824']['Number_coinitations'])
print(anova_4162)
```
